```python
import os  # For checking file existence and file operations
import random  # For generating random UIDs
import json  # For handling JSON file operations
from neo4j import GraphDatabase  # For connecting to and interacting with Neo4j
from neo4j.exceptions import Neo4jError  # For handling Neo4j-related exceptions


# Neo4j connection details
uri = "bolt://localhost:7687"
user = "neo4j"
password = "12345678"


class Neo4jConnection:
    def __init__(self, uri, user, password):
        self.driver = GraphDatabase.driver(uri, auth=(user, password))
        print("Connected to Neo4j.")

    def close(self):
        self.driver.close()
        print("Connection to Neo4j closed.")

    def generate_unique_uid(self, session):
        """Generates a unique 10-digit UID."""
        while True:
            uid = ''.join(random.choices('0123456789', k=10))  # Generate a random 10-digit number
            # Check if UID already exists
            result = session.run("MATCH (n) WHERE n.UID = $uid RETURN n LIMIT 1", uid=uid)
            if not result.single():  # If no node with this UID exists
                return uid

    def clear_database(self):
        """Clears the Neo4j database."""
        with self.driver.session(database="DocDatabase") as session:
            session.run("MATCH (n) DETACH DELETE n")
            print("Database cleared before importing documents.")

    def create_document(self, session, document_info):
        """Creates a Document node."""
        session.run("""
            CREATE (:Document {id: $id, title: $title, version: $version, date: $date})
        """, id=document_info['DocumentID'], title=document_info['title'], version=document_info['version'], date=document_info['date'])

    def create_passage(self, session, document_id, passage):
        """Creates a Passage node and links it to a Document or parent Passage."""
        raw_passage_id = passage.get('PassageID')
        
        # Check if PassageID exists
        if not raw_passage_id:
            print(f"Skipping Passage: Missing PassageID in document {document_id}")
            return None
        
        # Clean PassageID directly: Remove last character if it's a dot
        passage_id = str(raw_passage_id).replace('..', '.').strip()
        if passage_id.endswith('.'):
            passage_id = passage_id[:-1]  # Remove trailing dot
            #print(f"Create Passage: Removed trailing dot from PassageID: '{raw_passage_id}' to '{passage_id}'")
        
        # Extract UID and text
        uid = passage.get('ID',passage.get('ContextID', ''))
        text = passage.get('Text', passage.get('Passage', ''))
        
        # Check if UID is valid
        if not uid:
            print(f"Skipping Passage: Missing UID in document {document_id}, PassageID: {passage_id}")
            return None
        
        # Create Passage node in Neo4j
        session.run("""
            CREATE (p:Passage {UID: $uid, passageID: $passageID, text: $text})
        """, uid=uid, passageID=passage_id, text=text)
        
        return uid

    def link_passage_to_parent(self, session, passages_by_id, passage_id, uid, document_id):
        """Links a Passage node to its closest parent or grandparent, or directly to the Document."""
        current_passage_id = passage_id
        parent_found = False
    
        while not parent_found and '.' in current_passage_id:
            parent_passage_id = '.'.join(current_passage_id.split('.')[:-1])  # Move one level up
            parent_uid = passages_by_id.get(parent_passage_id)
    
            if parent_uid:
                # Link to the parent node
                session.run("""
                    MATCH (parent:Passage {UID: $parent_uid}), (child:Passage {UID: $child_uid})
                    MERGE (parent)-[:PARENT_OF]->(child)
                """, parent_uid=parent_uid, child_uid=uid)
                parent_found = True
                #print(f"Linked passage {uid} to parent {parent_uid} (PassageID: {parent_passage_id})")
            else:
                print(f"Parent not found for passage {current_passage_id}, moving to higher level: {parent_passage_id}")
                current_passage_id = parent_passage_id  # Move further up the hierarchy
    
        if not parent_found:
            print(f"No parent or grandparent found for passage {passage_id}. Linking to document.")
            self.link_passage_to_document(session, document_id, uid)


    def link_passage_to_document(self, session, document_id, uid):
        """Links a Passage node to a Document if no parent is found."""
        session.run("""
            MATCH (doc:Document {id: $document_id}), (passage:Passage {UID: $uid})
            MERGE (doc)-[:CONTAINS]->(passage)
        """, document_id=document_id, uid=uid)

    def create_obligations(self, session, passage_uid, obligations):
        """Creates Obligation nodes with unique UIDs and links them to a Passage."""
        if not obligations:
            return
        for obligation_text in obligations:
            heading, description = self.parse_obligation_text(obligation_text)
            uid = self.generate_unique_uid(session)  # Generate unique UID
            session.run("""
                MATCH (p:Passage {UID: $passage_uid})
                CREATE (o:Obligation {UID: $uid, heading: $heading, description: $description})
                MERGE (p)-[:HAS_OBLIGATION]->(o)
            """, passage_uid=passage_uid, uid=uid, heading=heading, description=description)
            
    def parse_obligation_text(self, text):
        lines = text.strip().split('\n')
        heading = lines[0].strip() if lines else ''
        description = '\n'.join(lines[1:]).strip() if len(lines) > 1 else ''
        return heading, description

    def create_named_entities(self, session, passage_uid, named_entities):
        """Creates NamedEntity nodes with unique UIDs and links them to a Passage."""
        for entity in named_entities:
            if isinstance(entity, dict):  # If it's a dictionary, extract 'Term' and 'Description'
                term = entity.get('ContextID', entity.get('ID', entity.get('Term', '')))
                description = entity.get('Description', entity.get('Meaning', ''))
            elif isinstance(entity, str):  # If it's a string, use it as the term and leave description empty
                term = entity
                description = ''
            else:
                print(f"Skipping invalid named entity: {entity}")
                continue
            
            uid = self.generate_unique_uid(session)  # Generate unique UID
            session.run("""
                MATCH (p:Passage {UID: $passage_uid})
                CREATE (e:NamedEntity {UID: $uid, term: $term, description: $description})
                MERGE (p)-[:HAS_NAMED_ENTITY]->(e)
            """, passage_uid=passage_uid, uid=uid, term=term, description=description)



    def create_defined_terms(self, session, passage_uid, defined_terms):
        """Creates DefinedTerm nodes with unique UIDs and links them to a Passage."""
        for term in defined_terms:
            if isinstance(term, dict):  # If it's a dictionary, extract 'Term' and 'Description'
                term_text = term.get('ContextID', term.get('ID', term.get('Term', '')))
                description = term.get('Description', term.get('Meaning', ''))
            elif isinstance(term, str):  # If it's a string, use it as the term and leave description empty
                term_text = term
                description = ''
            else:
                print(f"Skipping invalid defined term: {term}")
                continue
            
            uid = self.generate_unique_uid(session)  # Generate unique UID
            session.run("""
                MATCH (p:Passage {UID: $passage_uid})
                CREATE (d:DefinedTerm {UID: $uid, term: $term_text, description: $description})
                MERGE (p)-[:HAS_DEFINED_TERM]->(d)
            """, passage_uid=passage_uid, uid=uid, term_text=term_text, description=description)



    def parse_obligation_text(self, text):
        """Parses Obligation text into heading and description."""
        lines = text.strip().split('\n')
        heading = lines[0].strip() if lines else ''
        description = '\n'.join(lines[1:]).strip() if len(lines) > 1 else ''
        return heading, description

    def parse_processed_response(self, processed_response):
        obligations = []
        if processed_response:
            # Check for non-actionable responses
            non_actionable_phrases = [
                "Heading or Short Text: No analysis required.",
                "Explicit Chapter Applicability: No analysis required."
            ]
            if processed_response.strip() in non_actionable_phrases:
                return obligations  # Return empty list, no obligations to process

            lines = processed_response.split('\n')
            in_action_items = False
            obligation_text = ''
            for line in lines:
                line = line.strip()
                if '###' in line and 'Action Items' in line:
                    in_action_items = True
                    continue
                if in_action_items:
                    if line.startswith('By executing these action items') or line.startswith('**Note:**'):
                        # End of obligations
                        if obligation_text:
                            obligations.append(obligation_text.strip())
                        break
                    elif line.startswith(tuple(f"{i}." for i in range(1, 21))):
                        # Start of a new obligation
                        if obligation_text:
                            obligations.append(obligation_text.strip())
                        obligation_text = line
                    else:
                        obligation_text += '\n' + line
            if obligation_text:
                obligations.append(obligation_text.strip())
        return obligations

    def process_document(self, document_info, data):
        """Processes a document and its passages."""
        with self.driver.session(database="DocDatabase") as session:
            #print(f"Creating Document node for: {document_info['title']} (ID: {document_info['DocumentID']})")
            self.create_document(session, document_info)
            passages_by_id = {}
    
            for passage in data:
                # Ensure PassageID is a string and handle missing values
                raw_passage_id = passage.get('PassageID', passage.get('ID', ''))
                if raw_passage_id is None:
                    print(f"Skipping passage: Missing PassageID in document {document_info['DocumentID']}")
                    continue
                
                # Clean PassageID: Remove last character if it's a dot
                passage_id = str(raw_passage_id).replace('..', '.').strip()
                if passage_id.endswith('.'):
                    passage_id = passage_id[:-1]  # Remove trailing dot
                    #print(f"Process Document: Removed trailing dot from PassageID: '{raw_passage_id}' to '{passage_id}'")
    
                if not passage_id:
                    print(f"Skipping passage: Empty or invalid PassageID in document {document_info['DocumentID']}")
                    continue
    
                uid = self.create_passage(session, document_info['DocumentID'], passage)
                if not uid:
                    print(f"Skipped passage in document {document_info['title']}: Missing UID or PassageID.")
                    continue
    
                self.link_passage_to_parent(session, passages_by_id, passage_id, uid, document_info['DocumentID'])
                passages_by_id[passage_id] = uid
    
                # Create obligations, named entities, and defined terms
                try:
                    obligations = passage.get('Obligations') or self.parse_processed_response(passage.get('ProcessedResponse', ''))
                    if obligations:
                        self.create_obligations(session, uid, obligations)

                except Exception as e:
                    print(f"Error creating obligations for passage {uid}: {e}")
    
                try:
                    self.create_named_entities(session, uid, passage.get('NamedEntities', []))
                except Exception as e:
                    print(f"Error creating named entities for passage {uid}: {e}")
    
                try:
                    self.create_defined_terms(session, uid, passage.get('DefinedTerms', []))
                except Exception as e:
                    print(f"Error creating defined terms for passage {uid}: {e}")

if __name__ == "__main__":
    # Initialize Neo4j connection
    conn = Neo4jConnection(uri, user, password)
    conn.clear_database()

    for document_info in documents:
        print(f"Processing document: {document_info['title']} (ID: {document_info['DocumentID']})")
        
        # Check if the JSON file exists
        if not os.path.exists(document_info['json_file_path']):
            print(f"File not found: {document_info['json_file_path']}")
            continue

        # Load JSON data
        try:
            with open(document_info['json_file_path'], 'r', encoding='utf-8') as file:
                data = json.load(file)
        except json.JSONDecodeError as e:
            print(f"Failed to parse JSON file: {document_info['json_file_path']}. Error: {e}")
            continue

        # Process the document
        try:
            conn.process_document(document_info, data)
        except Exception as e:
            print(f"Error while processing document {document_info['title']} (ID: {document_info['DocumentID']}): {e}")

    # Close the Neo4j connection
    conn.close()

```

```python
import json
import os
from neo4j import GraphDatabase

# Neo4j connection details
uri = "bolt://localhost:7687"
user = "neo4j"
password = "12345678"

class Neo4jVerifier:
    def __init__(self, uri, user, password):
        self.driver = GraphDatabase.driver(uri, auth=(user, password))
        print("Connected to Neo4j.")

    def close(self):
        self.driver.close()
        print("Connection to Neo4j closed.")

    def get_passage_count(self):
        """Counts the total number of Passage nodes in the database."""
        with self.driver.session(database="DocDatabase") as session:
            result = session.run("MATCH (p:Passage) RETURN count(p) AS passage_count")
            return result.single()["passage_count"]

def count_passages_in_json(documents):
    """Counts the total number of passages in all JSON files."""
    total_passages = 0
    for doc in documents:
        json_file_path = doc.get("json_file_path")
        if not os.path.exists(json_file_path):
            print(f"File not found: {json_file_path}")
            continue

        try:
            with open(json_file_path, "r", encoding="utf-8") as file:
                data = json.load(file)
                passage_count = len(data)
                total_passages += passage_count
                print(f"File: {json_file_path}, Passages: {passage_count}")
        except json.JSONDecodeError as e:
            print(f"Error reading JSON file {json_file_path}: {e}")

    return total_passages

if __name__ == "__main__":

    # Count passages in JSON files
    print("Counting passages in JSON files...")
    total_passages_json = count_passages_in_json(documents)
    print(f"Total passages in JSON files: {total_passages_json}")

    # Count Passage nodes in Neo4j
    print("\nCounting Passage nodes in the Neo4j database...")
    verifier = Neo4jVerifier(uri, user, password)
    total_passages_neo4j = verifier.get_passage_count()
    print(f"Total Passage nodes in Neo4j: {total_passages_neo4j}")

    # Compare counts
    if total_passages_json == total_passages_neo4j:
        print("\n✅ The passage counts match between the JSON files and the Neo4j database!")
    else:
        print("\n❌ The passage counts do not match!")
        print(f"JSON Total: {total_passages_json}, Neo4j Total: {total_passages_neo4j}")

    # Close Neo4j connection
    verifier.close()
```
